```python
A1_height= [ list(df.loc["A1", "Day {}".format(i+1)].mean())[0] for i in range(11)]
A1_leaves= [ list(df.loc["A1", "Day {}".format(i+1)].mean())[1] for i in range(11)]
A1_stems= [ list(df.loc["A1", "Day {}".format(i+1)].mean())[2] for i in range(11)]


plt.style.use('seaborn-whitegrid')

fig, ax1 = plt.subplots(figsize=(12, 7))


days = np.arange(1,12)
x_indexes= np.arange(len(days))
width= 0.25


ax1.set_xlabel('Days')
plot1 = ax1.plot(x_indexes-width,A1_height,color="#3376bd",linestyle='--',marker='o',label= "Plant Height")#
ax1.set_ylabel("Height (cm)")

ax2 = ax1.twinx() 
plot2= ax2.bar(x_indexes,A1_leaves,width= width,color="#edad49",label= "Numbers of Leaves")
ax2.set_ylabel("Count") 


plot3=ax2.bar(x_indexes+width,A1_stems,width= width,color="#00798c",linestyle='--',label= "Numbers of Stems")

plt.title("A1", fontweight=14)
#plt.text(-0.5, -2, 'Sine wave', fontsize = 14)
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

plt.title("Plot A1", fontsize = 16)
ax1.grid(axis='x')

ax1.legend(loc= "upper left",fontsize=12)
ax2.legend(bbox_to_anchor=(0.1,0.84), loc='lower center',fontsize=12)
plt.xticks(ticks= x_indexes, labels= days)
plt.tight_layout()
```
